```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn import metrics
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, Normalizer ,RobustScaler ,PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, LassoCV
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error ,r2_score


import warnings
warnings.filterwarnings('ignore')
```

```python
header = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

df = pd.read_csv('housing.csv', header=None, names=header, delim_whitespace=True)

df
```

```python
# Define the data/predictors
x = df.drop("MEDV" , axis=1)


# Define the target variable (dependent variable)
y = df.MEDV

df_x = pd.DataFrame(x)
df_y = pd.DataFrame(y)

df_x
```

```python
# create training and testing vars
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.1, random_state=101)

print (x_train.shape, y_train.shape)

print (x_test.shape, y_test.shape)
```

```python
def polynomial_features(degree, train, test):
    
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    train_poly = poly.fit_transform(train)
    test_poly = poly.transform(test)
    
    return train_poly, test_poly
```

```python
def polynomial_regression(degree, cv, train, test):
    
    train_poly, test_poly = polynomial_features(degree, train, test)
    
    
    print('Train data and test data shapes are: ', train_poly.shape, test_poly.shape)
    print('degree = ', degree)
    print('CV = ', cv)
    
    
    
    model = LinearRegression()
    model_gscv = GridSearchCV(model,param_grid  = {'fit_intercept': [True]}, cv=cv,
                              verbose=3, scoring='neg_mean_absolute_error')
    model_gscv.fit(train_poly, y_train)
    

    scores = []
    best_cv = 0
    for i in range(cv):
        scores.append(model_gscv.cv_results_['split'+str(i)+'_test_score'][0])
        best_cv = scores.index(max(scores))    
    
    best_score = -max(scores)
    
    print('best_score = ', best_score)
    print('best_cv = ', best_cv+1)
    
    
    print('----------------------------------------------------------------------------------')
    return 
```

Try Different Cross-Validation Splits and Degrees to know which degree and Cross-Validation Splits to get best score

```python
for i in range(1,5):
    
    for cv in range(5,11):
    
        polynomial_regression(i, cv, x_train, x_test)
```

Best score overall is = 1.5939 at CV(5/10), degree = 2

Evaluate our model on the test data

```python
x_train_poly, x_test_poly = polynomial_features(2 , x_train, x_test)
print('Train data and test data shapes are: ', x_train_poly.shape, x_test_poly.shape)
```

```python
model = LinearRegression().fit(x_train_poly, y_train)

y_pred_train = model.predict(x_train_poly)

y_pred_test = model.predict(x_test_poly)


print('train data score = ', mean_absolute_error(y_pred_train, y_train))
print('test data score = ', mean_absolute_error(y_pred_test, y_test))


print("R-Square Value",r2_score(y_test,y_pred_test))
print ("MSE : ",metrics.mean_squared_error(y_test, y_pred_test))
print ("RMSE : ",np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
```

## Ridge regression

alpha: must be positive, increase for more regularization

```python
def range_with_floats(start, stop, step):
    while stop > start:
        yield start
        start += step
```

```python
for a in range_with_floats(0, 0.1, 0.01):
    
    ridgereg = Ridge(alpha=a, normalize=True)
    ridgereg.fit(x_train_poly, y_train)

    y_pred_train = ridgereg.predict(x_train_poly)
    y_pred_test = ridgereg.predict(x_test_poly)
    
    print('alpha = ', a)
    print('train data score = ', mean_absolute_error(y_pred_train, y_train))
    print('test data score = ', mean_absolute_error(y_pred_test, y_test))
    
    
    print("R-Square Value",r2_score(y_test,y_pred_test))
    print ("MSE : ",metrics.mean_squared_error(y_test, y_pred_test))
    print ("RMSE: ",np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
    print('----------------------------------------------------------------------------------')
```

```python
# create an array of alpha values
alpha_range = [*range_with_floats(0.01, 0.1, 0.01)]

# select the best alpha with RidgeCV
ridgeregcv = RidgeCV(alphas=alpha_range, normalize=True, scoring='neg_mean_absolute_error')
ridgeregcv.fit(x_train_poly, y_train)
print('ridgeregcv.alpha = ' , ridgeregcv.alpha_)


# predict method uses the best alpha value
y_pred_train = ridgeregcv.predict(x_train_poly)
y_pred_test = ridgeregcv.predict(x_test_poly)
# calculate R^2 value, MAE, MSE, RMSE

print('train data score = ', mean_absolute_error(y_pred_train, y_train))
print('test data score = ', mean_absolute_error(y_pred_test, y_test))


print("R-Square Value",r2_score(y_test,y_pred_test))
print ("mean_absolute_error :",metrics.mean_absolute_error(y_test, y_pred_test))
print ("MSE : ",metrics.mean_squared_error(y_test, y_pred_test))
print ("RMSE : ",np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
```
